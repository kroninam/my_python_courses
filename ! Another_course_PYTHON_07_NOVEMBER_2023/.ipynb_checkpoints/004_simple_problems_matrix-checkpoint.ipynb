# Задача №4

```python
Напишите программу, которая выводит количество элементов квадратной матрицы 
в каждой строке, больших среднего арифметического элементов данной строки.

Формат входных данных 
На вход программе подаётся натуральное число 
n — количество строк и столбцов в матрице, 
затем элементы матрицы (целые числа) построчно через пробел.

Формат выходных данных
Программа должна вывести
n чисел — для каждой строки количество элементов матрицы, 
больших среднего арифметического элементов данной строки.
```

```python
n = int(input())

for i in range(n):
    a = input().split()
    mean = 0
    for j in a:
        mean += int(j)
    mean = mean/len(a)
    
    rez = 0
    for k in a:
        if int(k) > mean:
            rez += 1
    print(rez)
            
```

```python
Первая четверть это i < j and i < n - 1 - j
Вторая четверть это i < j and i > n - 1 - j
Третья четверть это i > j and i > n - 1 - j
Четвертая четверть это i > j and i < n - 1 - j
```

```python
Квадратная матрица разбивается на четыре четверти, ограниченные главной и побочной диагоналями: верхнюю, нижнюю, левую и правую.
```

```python
Напишите программу, которая вычисляет сумму элементов: верхней четверти; правой четверти; нижней четверти; левой четверти.

Формат входных данных
На вход программе подаётся натуральное число 
n — количество строк и столбцов в матрице, затем элементы матрицы (целые числа) построчно через пробел.

Формат выходных данных
Программа должна вывести текст в соответствии с условием задачи.

Примечание. Элементы диагоналей не учитываются.
```

```python
n = int(input())

lst = []

for i in range(n):
    tmp = list(map(int, input().split()))
    lst.append(tmp)
```

```python
lst
```

```python
ch_1 = []
ch_2 = []
ch_3 = []
ch_4 = []

for i in range(n):
    for j in range(len(lst[i])):
        if i < j and i < n - 1 - j:
            ch_1.append((lst[i][j]))
            
        if i < j and i > n - 1 - j:
            ch_2.append((lst[i][j]))
            
        if i > j and i > n - 1 - j:
            ch_3.append((lst[i][j]))
            
        if i > j and i < n - 1 - j:
            ch_4.append((lst[i][j]))

            
    
```

```python
print('Верхняя четверть:', sum(ch_1))
print('Правая четверть:', sum(ch_2))
print('Нижняя четверть:', sum(ch_3))
print('Левая четверть:', sum(ch_4))
```

# Задача №16

```python
На вход программе подаются два натуральных числа 
n и m. Напишите программу для создания матрицы размером 
n×m, заполнив её символами . и * в шахматном порядке.

В левом верхнем углу должна стоять точка. Выведите полученную матрицу на экран, разделяя элементы пробелами.

Формат входных данных
На вход программе на одной строке через пробел подаются два натуральных числа 
n и m — количество строк и столбцов в матрице.

Формат выходных данных
Программа должна вывести матрицу, описанную в условии задачи.
```

```python
n, m = map(int, input().split())
```

```python
lst = []

for i in range(n):

    if i%2 == 0:
        if m == 1:
            lst.append(['.'])
        else:
            lst.append(['.',  '*'] * (m//2))
    else:
        if m == 1:
            lst.append(['*'])
        else:
            lst.append(['*',  '.'] * (m//2))
            
```

```python
lst
```

```python
for i in lst:
    for j in i:
        print(j, end = ' ')
    print()
```

# Задача №17

```python
На вход программе подается натуральное число 
n. Напишите программу, которая создает матрицу размером 

n×n и заполняет её по следующему правилу:

числа на побочной диагонали равны 1;
числа, стоящие выше этой диагонали, равны 0;
числа, стоящие ниже этой диагонали, равны 2.

Полученную матрицу выведите на экран. Числа в строке разделяйте одним пробелом.

Формат входных данных
На вход программе подается натуральное число 
n — количество строк и столбцов в матрице.

Формат выходных данных
Программа должна вывести матрицу в соответствии с условием задачи.
```

```python
n = int(input())
```

```python
lst = []

for i in range(n):
    lst.append([0]*n)
```

```python
lst
```

```python
for i in range(n):
    for j in range(n):
        if j == n - 1 - i:
            lst[i][j] = 1
            
        if j > n - 1 - i:
            lst[i][j] = 2
```

```python
lst
```

```python
for i in range(n):
    for j in range (n):
        print(lst[i][j], end = ' ')
    print()
```

```python
Заполнение 1
На вход программе подаются два натуральных числа 
n и m. Напишите программу, которая создает матрицу размером 

n×m и заполняет её числами от 

1 до 

n⋅m в соответствии с образцом.

Формат входных данных
На вход программе на одной строке подаются два натуральных числа 
n и m — количество строк и столбцов в матрице.

Формат выходных данных
Программа должна вывести матрицу в соответствии с образцом.

Примечание. Для вывода элементов матрицы как в примерах отводите ровно 
3 символа на каждый элемент. Для этого используйте строковый метод ljust().
```

```python
n, m = map(int, input().split())
```

```python
lst = []
for i in range(n):
    lst.append([j+1 for j in range(m)])    
```

```python
lst
```

```python
lst = []
x = 0

for i in range(n):
    tmp = []
    for j in range (m):        
        tmp.append(x+1)
        x += 1
    lst.append(tmp)
```

```python
for i in range(n):
    for j in range(m):
        print(str(lst[i][j]).ljust(3), end = ' ')
    print()
```
