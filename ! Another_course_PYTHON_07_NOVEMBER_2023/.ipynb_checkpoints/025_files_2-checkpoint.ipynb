```python
with open('output.txt', 'w') as file1:
    

    with open('files/logfile.txt', encoding = 'utf-8') as file:
        lst = file.readlines()
        
        for i in range(len(lst)):
            tim = int(lst[i].strip().split(',')[2].strip().replace(':', '')) - int(lst[i].strip().split(',')[1].strip().replace(':', ''))
            if tim >= 100:
                s = lst[i].split(',')[0] + '\n'
                file1.write(s)
```
