```python
С помощью библиотеки Matplotlib постройте график функции 

y=∣cos9.8x−cos10x∣ на отрезке x от 0 до 5 и определите количество максимумов (пиков) графика.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 1000)
x
```

```python
y = []
for i in x:
    y.append(np.abs(  np.cos(9.8*i) - np.cos(10*i)   ))
    
```

```python
y
```

```python
plt.plot(x, y)
```
