```python
with open ('files/files.txt') as file:
    lst = file.readlines()
```

```python
# lst
```

```python
d = {}
for i in lst:
    stroka = i.strip().split()
    # print(stroka[0].split('.')[0])
    d[str(stroka[0].split('.')[0])] = [stroka[0].split('.')[1], stroka[1], stroka[2]]
```

```python
# d
```

```python
dd = {}
for k,w in d.items():
    if w[0] not in dd:
        dd[w[0]] = [(k, w[1], w[2])]
    else:
        dd[w[0]] += [(k, w[1], w[2])]
```

```python
# dd
```

```python
for k, w in sorted(dd.items()):
            
    summary_b = 0
    for i in sorted(w):
        print(i[0] + '.' + k)
        
        if i[2] == 'GB':
            summary_b += int(i[1]) * 1024 * 1024 * 1024
        if i[2] == 'MB':
            summary_b += int(i[1]) * 1024 * 1024
        if i[2] == 'KB':
            summary_b += int(i[1]) * 1024
        if i[2] == 'B':
            summary_b += int(i[1])

    print('----------')

    if any('GB' in j for j in w):
        print('Summary:', round(summary_b / (1024* 1024 * 1024)), 'GB')
        
    elif any('MB' in j for j in w):

        if (summary_b // (1024* 1024)) < 1024:
            
            print('Summary:', round(summary_b / (1024 * 1024)), 'MB')
        else:
            print('Summary:', round((summary_b / (1024* 1024)) / (1024)), 'GB')            
        
    
    elif any('KB' in j for j in w):
        print('Summary:', round(summary_b / (1024)), 'KB')
        

    else:
        print('Summary:', round(summary_b), 'KB')
    print()
        
    
        
```
