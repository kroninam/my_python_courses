```python
from datetime import date, time, datetime

dates = [date(1793, 8, 23), date(1410, 3, 11), date(804, 11, 12), date(632, 6, 4),
         date(295, 1, 23), date(327, 8, 24), date(167, 4, 16), date(229, 1, 24), 
         date(1239, 2, 5), date(1957, 7, 14), date(197, 8, 24), date(479, 9, 6)]

times = [time(7, 33, 27), time(21, 2, 10), time(17, 20, 47), time(20, 8, 59), 
         time(12, 42, 56), time(15, 9, 57), time(17, 47, 9), time(9, 40, 2), 
         time(11, 47, 1), time(17, 27, 10), time(17, 55, 40), time(9, 14, 9)]

lst = []
for i, j in zip(dates, times):
    lst.append(datetime.combine(i, j))
print(*sorted(lst, key = lambda x: x.second), sep = '\n')
```

```python
from datetime import datetime

with open('files/cosmos_file/diary.txt', encoding = 'UTF-8') as file:
    
    lst = file.readlines()
    
```

```python
d = {}
for i in lst:
    print(i.strip())

    try:
        data_time = datetime.strptime(i.strip(), '%d.%m.%Y; %H:%M')
        d[data_time] = []
        print(d)
    except:
        d[data_time] += [i]
```

```python
for k, w in sorted(d.items()):
    kk = datetime.strftime(k, '%d.%m.%Y; %H:%M')
    print(str(kk))
    for i in w:
        if i != '\n':
            print(i.strip())
    print()
```

```python
with open('files/cosmos_file/ans.txt', 'w') as f:
    
    for k, w in sorted(d.items()):
        kk = datetime.strftime(k, '%d.%m.%Y; %H:%M')
        f.write(str(kk))
        f.write('\n')
        for i in w:
            if i != '\n':
                f.write(i)
        f.write('\n')
```

```python
def is_available_date(booked_dates, date_for_booking):

    if len(date_for_booking.split('-')) == 1:
    
        for i in booked_dates:
            if i == date_for_booking:
                return False
            elif len(i.split('-')) == 1:
                if date_for_booking == i[0]:
                    return False
    
            elif len(i.split('-')) == 2:
                
                if datetime.strptime(date_for_booking, '%d.%m.%Y') >= datetime.strptime(i.split('-')[0], '%d.%m.%Y') and datetime.strptime(date_for_booking, '%d.%m.%Y') <= datetime.strptime(i.split('-')[1], '%d.%m.%Y'):
                    return False
    
        return True


    else:
        
        for i in booked_dates:
            
            if i == date_for_booking:
                return False
            elif len(i.split('-')) == 1:
                if datetime.strptime(i.split('-')[0], '%d.%m.%Y') >= datetime.strptime(date_for_booking.split('-')[0], '%d.%m.%Y') and datetime.strptime(i.split('-')[0], '%d.%m.%Y') <= datetime.strptime(date_for_booking.split('-')[1], '%d.%m.%Y'):
                    return False
            else:
                if datetime.strptime(date_for_booking.split('-')[1], '%d.%m.%Y') == datetime.strptime(i.split('-')[0], '%d.%m.%Y'):
                    return False
                elif datetime.strptime(date_for_booking.split('-')[0], '%d.%m.%Y') == datetime.strptime(i.split('-')[1], '%d.%m.%Y'):
                    return False
                elif datetime.strptime(date_for_booking.split('-')[0], '%d.%m.%Y') == datetime.strptime(i.split('-')[0], '%d.%m.%Y'):
                    return False
                elif datetime.strptime(date_for_booking.split('-')[1], '%d.%m.%Y') == datetime.strptime(i.split('-')[1], '%d.%m.%Y'):
                    return False
                elif datetime.strptime(i.split('-')[0], '%d.%m.%Y') >= datetime.strptime(date_for_booking.split('-')[0], '%d.%m.%Y'):
                    if datetime.strptime(i.split('-')[1], '%d.%m.%Y') <= datetime.strptime(date_for_booking.split('-')[1], '%d.%m.%Y'):
                        return False
                elif datetime.strptime(i.split('-')[0], '%d.%m.%Y') <= datetime.strptime(date_for_booking.split('-')[0], '%d.%m.%Y'):
                    if datetime.strptime(i.split('-')[1], '%d.%m.%Y') >= datetime.strptime(date_for_booking.split('-')[1], '%d.%m.%Y'):
                        return False

        return True
```
