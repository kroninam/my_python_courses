```python
import csv

with open ('files_csv/salary_data.csv', encoding = 'utf-8') as file:
    d = {}
    for i in (list(file)[1:]):
        if i.strip('\n').split(';')[0] not in d:
            d[i.strip('\n').split(';')[0]] = [i.strip('\n').split(';')[1]]
        else:
            d[i.strip('\n').split(';')[0]] += [i.strip('\n').split(';')[1]]
        
```

```python
res = []
for k,w in d.items():
    sum_zp = 0
    for i in w:
        sum_zp += int(i)
    mean_zp = sum_zp/len(w)
    res.append((k, mean_zp))
    
```

```python
for i in sorted(res, key = lambda x: (x[1], x[0])):
    print(i[0])
```

```python
import csv

def csv_columns(filename):
    with open (filename, encoding = 'utf-8') as file:
        reader = file.readline()
        data = file.readlines()


    keeys = reader.strip().split(';')
    d = {}


    for i in range(len(keeys)):
        
        lst = []
        for row in data:
            lst.append(row.strip().split(';')[i])
    
        d[keeys[i]] = lst

    return d
```

```python
with open('files_csv/data.csv', encoding = 'UTF-8') as file:
    file.readline()
    lst = file.readlines()
```

```python
lst
```

```python
d = {}

for row in lst:
    
    key = row.strip().split(',')[2].split('@')[1]
    print(key)
    if key not in d:
        d[key] = 1
    else:
        d[key] += 1
```

```python
d
```

```python
with open('files_csv/domain_usage.csv', 'w', encoding = 'UTF-8') as file:
    file.write('domain,count\n')
    for k, w in sorted(d.items(), key = lambda x: (x[1], x[0])):
        file.write(f'{k},{w}\n')
```

```python
with open ('files_csv/wifi.csv', encoding = 'utf-8') as file:
    file.readline()
    lst = file.readlines()
```

```python
lst
```

```python
d = {}
for row in lst:
    #print(row.strip().split(';'))
    key = row.strip().split(';')[1]
    val = int(row.strip().split(';')[-1])
    if key not in d:
        d[key] = val
    else:
        d[key] += val
```

```python
for k, w in sorted(d.items(), key = lambda x: (x[1], x[0]), reverse = (True, False)):
    print(f'{k}: {w}')
```

```python
with open('files_csv/name_log.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)

    d = {}
    
    for i in reader:
        # print(i['email'])

        if i['email'] not in d:
            d[i['email']] = [(i['username'], i['dtime'])]
        else:
            d[i['email']] += [(i['username'], i['dtime'])]
            
       
```

```python
from datetime import datetime

'07/11/2021 01:07'

datetime.strptime('07/11/2021 01:07', '%d/%m/%Y %H:%M')
```

```python
res = 'username,email,dtime\n'

for k, w in sorted(d.items(), key = lambda x: (x[0], x[1])):

    for i in sorted(w, 
                    key = lambda x: datetime.strptime(x[1], '%d/%m/%Y %H:%M'), reverse = True
                   ):
        # print(i[0], k, i[1])
        res += f'{i[0]},{k},{i[1]}\n'
        break
    # break
```

```python
for i in res.split('\n'):
    print(i)
```

```python
with open('files_csv/new_name_log.csv', 'w') as file:
    file.write(res)    
```
